```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def main():
    # Load dataset
    print("Loading Iris dataset...")
    df = sns.load_dataset("iris")

    print("\n--- Basic Information ---")
    # Print shape
    print(f"Shape: {df.shape}")

    # Print column names
    print(f"Columns: {df.columns.tolist()}")

    # Print first few rows
    print("\nFirst 5 rows:")
    print(df.head())

    print("\n--- Summary Statistics ---")
    # info() prints to stdout, capture or just call it
    print("\nInfo:")
    df.info()

    # describe()
    print("\nDescribe:")
    print(df.describe())

    print("\n--- Generating Visualizations ---")

    # 1. Scatter plot
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
    plt.title('Sepal Length vs Sepal Width')
    plt.savefig('scatter_plot.png')
    plt.show()
    plt.close()

    # Pairplot for scatter plots of all features relationships
    print("Generating pairplot...")
    sns.pairplot(df, hue='species')
    plt.savefig('pairplot.png')
    plt.show()
    plt.close()

    # 2. Histograms
    print("Generating histograms...")
    plt.figure(figsize=(10, 8))
    for i, feature in enumerate(['sepal_length', 'sepal_width', 'petal_length', 'petal_width']):
        plt.subplot(2, 2, i+1)
        sns.histplot(data=df, x=feature, kde=True, hue='species')
        plt.title(f'Histogram of {feature}')
    plt.tight_layout()
    plt.savefig('histograms.png')
    plt.show()
    plt.close()

    # 3. Box plots
    print("Generating boxplots...")
    plt.figure(figsize=(10, 8))
    for i, feature in enumerate(['sepal_length', 'sepal_width', 'petal_length', 'petal_width']):
        plt.subplot(2, 2, i+1)
        sns.boxplot(data=df, x='species', y=feature)
        plt.title(f'Boxplot of {feature}')
    plt.tight_layout()
    plt.savefig('boxplots.png')
    plt.show()
    plt.close()

    print("\nVisualizations saved successfully!")

if __name__ == "__main__":
    main()
```
